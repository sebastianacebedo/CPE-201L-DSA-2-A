```python
def traverse_array(arr):
    """Function to traverse and print all elements in the array"""
    print("Traversing the array:")
    for i, element in enumerate(arr):
        print(f"Index {i}: {element}")

def reverse_array(arr):
    print("Reversing all elements")
    arr.reverse()
    for i, element in enumerate(arr):
        print(f"Index {i}: {element}")

def sort_array(arr):
    print("Sorting all elements")
    arr.sort()
    for i, element in enumerate(arr):
        print(f"Index {i}: {element}")

def insert_array(arr):
    print("Inserting an element")
    index, new_element = input("Enter the index and element you want to insert: ").split()
    arr.insert(int(index), int(new_element))
    for i, element in enumerate(arr):
        print(f"Index {i}: {element}")

def delete_array(arr):
    print("Deleting an element")
    index = input("Enter the index of the element you want to delete: ")
    del arr[int(index)]
    for i, element in enumerate(arr):
        print(f"Index {i}: {element}")

def modify_array(arr):
    print("Modifying an element")
    index, new_element = input("Enter the index and new element you want to modify: ").split()
    arr[int(index)] = int(new_element)
    for i, element in enumerate(arr):
        print(f"Index {i}: {element}")

def main():
    arr = [10, 20, 30, 40, 50]
    print("Original array:", arr)

    while True:
        print("\nMenu:")
        print("1. Traverse array")
        print("2. Reverse array")
        print("3. Sort array")
        print("4. Insert element")
        print("5. Delete element")
        print("6. Modify element")
        print("7. Exit")
        choice = input("\nEnter your choice (1-7): ")

        if choice == '1':
            traverse_array(arr)
        elif choice == '2':
            reverse_array(arr)
        elif choice == '3':
            sort_array(arr)
        elif choice == '4':
            insert_array(arr)
        elif choice == '5':
            delete_array(arr)
        elif choice == '6':
            modify_array(arr)
        elif choice == '7':
            print("Exiting the program.")
            break
        else:
            print("Invalid choice. Please enter a number between 1 - 7.")

if __name__ == "__main__":
    main()
```
